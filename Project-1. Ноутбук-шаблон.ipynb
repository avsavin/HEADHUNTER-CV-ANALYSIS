```python
#ваш код здесь

def education(val):
    # воспользуемся словарем для преобразования значений
    translate  = dict({
        "['Высшее', 'образование']" : "высшее",  
        "['Среднее', 'специальное']" : "неоконченное высшее",
        "['Неоконченное', 'высшее']" : "среднее специальное",
        "['Среднее', 'образование']" : "среднее"
    })
    key = str(val.split()[:2]) # первых двух слов достаточно для формирования ключа
    return translate[key]


df['Образование'] = df['Образование и ВУЗ'].apply(education)

df = df.drop('Образование и ВУЗ', axis=1)
```

* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь

def work_experience (val):
    val_list = []
    years = int(0)
    months = int(0)
    total_months = np.NaN
    if val is not np.NaN :
        val_list = val.split()
    if len(val_list) <= 3 : 
        # недостаточно данных в формате
        return np.NaN
    if val_list[0] == 'Опыт' and val_list[1] == 'работы':   
        # строка содержит данные об опыте работы
        if val_list[3] in ['год','года','лет'] : 
            # если в четверой позиции "год", "года" или  "лет", то третий элемент списка должет содержать колечество лет
            years = int(val_list[2])
            if len(val_list) >= 6:
                if val_list[5] in ['месяц','месяцев','месяца'] : 
                    # если указано и количество лет и количество месяцев, дополнительно колечество месяцев будет в пятой позиции
                    months = int(val_list[4])
        elif val_list[3] in ['месяц','месяцев','месяца'] : 
            # в этом случае третий элемент - количество месяцев
            months = int(val_list[2])
        total_months = 12 * years + months
    return total_months

df["Опыт работы (месяц)"] = df['Опыт работы'].apply(work_experience)
```

```python
#ваш код здесь

def base_and_mobility(value):
    
    million_cities = list(['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 
                      'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград'])
    city = np.NaN
    relocation = np.NaN
    trips = False 
    # выделяем признаки, разделенные запятыми
    value_splitted = value.split(sep=',') 

    for i in range(len(value_splitted)):
        if (i == 0) :
            # Названия городов миллионников всегда в первой позиции
            city = value_splitted[i].strip()
            if city in  ["Москва", "Санкт-Петербург"]:
                pass
            else:
                city = 'город-миллионник' if city in million_cities else 'другие'
        else:
            # выделяем признаки готовности к переезду и/или командировкам отбросив информацию после скобок
           
            token = value_splitted[i].split(sep='(')[0].strip()
            
            if token in [
                'готов к командир',
                'готов к команд',
                'готов к командировкам',
                'готов к редким командировкам',
                'готова к командировкам',
                'готова к редким командировкам'
                ] : 
                trips = True
            elif token in [
                'готов к переезду',
                'хочу переехать',
                'готова к переезду'          
                ]:
                relocation = True
            elif token in [
                'не готов к переезду',
                'не готова к переезду'
                ]:
                relocation = False

    return pd.Series([city, relocation, trips], index= ['Город','переезд','командировки'])

        
df[['Город','переезд','командировки']] = df['Город, переезд, командировки'].apply(base_and_mobility)
df = df.drop('Город, переезд, командировки',axis=1)
print(round(df['Город'].value_counts(normalize=True)['Санкт-Петербург'] * 100))
print(round(df[
    df['переезд'] & df['командировки']
].shape[0] / df.shape[0] *100))
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь

# получим список всех возможных признаков из колонки Занятость 
lst = list((df['Занятость'].apply(lambda x : x.split(sep=','))))
occupancy_flags = list(set([item.lstrip() for sublist in lst for item in sublist ])) # в некоторых значениях были лидирующие нули

# для колонки График воспользуемся представленным в задании списком признаков
schedule_flags = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']

def occupancy(value):  
    flags = pd.Series(index=[occupancy_flags], dtype='object')
    for cat in occupancy_flags:
        flags[cat] = True if cat in value else False
    return flags

def schedule(value):
    flags = pd.Series(index=[schedule_flags], dtype='object')
    for cat in schedule_flags:
        flags[cat] = True if cat in value else False
    return flags

df[occupancy_flags] = df['Занятость'].apply(occupancy)
df[schedule_flags] = df['График'].apply(schedule)

print(df[df['проектная работа']&df['волонтерство']].shape[0])
print(df[df['вахтовый метод']&df['гибкий график']].shape[0])
```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь

# таблица преобразования кодов валюты исходных данных и пропорций
currency = pd.DataFrame({
    'грн.':	['UAH',	10,	'гривна'],
    'USD':	['USD',	1,	'доллар'],
    'EUR':	['EUR',	1,	'евро'],
    'бел.руб.': ['BYN',	1,	'белорусский рубль'],
    'KGS':	['KGS',	10,	'киргизский сом'],
    'сум':	['UZS',	10000,	'узбекский сум'],
    'AZN':	['AZN',	1,	'азербайджанский манат'],
    'KZT':	['KZT',	100,	'казахстанский тенге']},
    index = ['code','proportion','description']
)
    
# получим данные обменных курсов
currency_exchange = pd.read_csv('data/ExchangeRates.csv')
# преобразуем форматы даты
currency_exchange['date'] = pd.to_datetime(currency_exchange['date'], dayfirst=True )
df['date'] = pd.to_datetime(df['Обновление резюме'], dayfirst = True, ).dt.date
# конвертируем зарплату в рубли
def convert_amount(value):
    
    amount, currency_code, date = np.float64(value['ЗП'].split()[0]), value['ЗП'].split()[1], value['date']
    rate = 0
    if currency_code != 'руб.':
        iso_code, proportion = currency[currency_code].code, currency[currency_code].proportion
        
        rate = currency_exchange[(currency_exchange['currency']==iso_code)&(currency_exchange['date']==pd.to_datetime(date))].iloc[0][4] 
       
        converted_amount = amount * rate / proportion
    else :
        converted_amount = amount 
    return converted_amount


df['Salary'] = df[['ЗП','date']].apply(convert_amount,axis=1)

print(df['Salary'].mean())

```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

`histplot.axvline(log_age.mean(), color='k', lw=2)`

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# ваш код здесь

log_age = np.log10(clean_df['Возраст'])
mu = log_age.mean()
sigma = log_age.std()
lower_bound = mu - 3 * sigma
upper_bound = mu + 4 * sigma

logplt = sns.histplot(data=clean_df, x= log_age, bins=20 )
logplt.set_title('Логарифмическое распределение возраста соискателей\n(за пределами красных пунктирных линий - выбросы)')
logplt.axvline(mu, color='k', lw=2);
logplt.axvline(mu-3*sigma, color='r', lw=2, linestyle = 'dotted');
logplt.axvline(mu+3*sigma, color='g', lw=2, linestyle = 'dotted');
logplt.axvline(mu+4*sigma, color='r', lw=2, linestyle = 'dotted');
    
outliers = log_age[ (log_age < lower_bound) | (log_age > upper_bound)]
cleaned = log_age[(log_age > lower_bound) & (log_age < upper_bound)]
  
print('Выбросы =', outliers.count(),'\n'
      'Возраст слева = ', round(10**lower_bound),
      ' Возраст справа = ', round(10**upper_bound)
      )

clean_df = clean_df.drop(index=outliers.index)
```

логарифмический
